```python
# Task 1
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
x_values = [1.0, 2.0, 3.0]
y_values = [2.0, 4.0, 1.0]

# Creating DataFrame
df = pd.DataFrame({'X': x_values, 'Y' : y_values})


# Plotting the data
plt.plot(df['X'], df['Y'])
plt.title('Sample Line Graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

# Setting limits to match the picture given in example
plt.xlim(left = 1.0)
plt.ylim(bottom = 1.0)
plt.ylim(top = 4.0)

# Displaying the plot
plt.show()
```

```python
# Task 2 
# Sample data for plotting
x_values = [10, 20, 30]
y_values1 = [40, 10, 30]
y_values2 = [20, 40, 10]

df1 = pd.DataFrame({'X': x_values, 'Y': y_values1})
df2 = pd.DataFrame({'X': x_values, 'Y': y_values2})

# Plotting two lines with different widths and colors
plt.plot(df1['X'], df1['Y'], label='Line 1 - width - 5', color='red', linewidth=5)
plt.plot(df2['X'], df2['Y'], label='Line 2 - width - 3', color='blue', linewidth=3)

# Adding legend
plt.legend()

# Adding title and axis labels to match picture in example
plt.title('Two Lines with Different Widths and Colors')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.xlim(left = 10)
plt.xlim(right = 30)
plt.ylim(bottom = 10)
plt.ylim(top = 40)


# Setting x-axis and y-axis intervals of 5
plt.xticks(range(10, 31, 5))
plt.yticks(range(10, 41, 5))

# Displaying the plot
plt.show()
```

```python
# Task 3 


# Sample data for plotting
# Specific points provided
x_values1 = [1, 4, 5, 6, 7]
y_values2 = [2, 6, 3, 6, 3]
x_values2 = [2, 3, 5, 7, 8]
y_values2 = [1, 4, 6, 8, 7]



# Plotting two lines with specified markers and styles
plt.plot(x_values1, y_values1, label='Line 1', color='red', linewidth=2, marker='o', linestyle='dotted') 
plt.plot(x_values1, y_values2, label='Line 2', color='blue', linewidth=2, marker='o', linestyle='dashdot')  

# Adding legend
plt.legend()

# Adding title and axis labels
plt.title('Two Lines with Different Markers')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.xlim(left = 1)
plt.xlim(right = 8)
plt.ylim(bottom = 1)
plt.ylim(top = 8)

# Displaying the plot
plt.show()
```

```python
# Task 4

# Given data
programming_languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

# Creating the horizontal bar chart
plt.figure(figsize=(10, 5))
bars = plt.barh(programming_languages, popularity, color='green')

# Adding title and axis labels
plt.xlabel('Popularity')
plt.title('Popularity of Programming Languages\nWorldwide, Oct 2017 compared to a year ago')



# Setting x-axis intervals and adding grid
plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.5', color='red')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')

# Customizing tick params
plt.tick_params(which='both', width=0.5)
plt.tick_params(which='major', length=7)
plt.tick_params(which='minor', length=4, color='black')

# Setting the limits for x-axis to match the graph in the image
plt.xlim(0, 25)

# Displaying the plot
plt.show()
```

```python
# Task 5 

# Given data
data = {
    'a': [2, 4, 6, 8, 10],
    'b': [4, 2, 4, 2, 4],
    'c': [8, 3, 7, 4, 3],
    'd': [5, 4, 7, 8, 3],
    'e': [6, 6, 8, 6, 2]
}

# Creating a DataFrame
df = pd.DataFrame(data, index=[2, 4, 6, 8, 10])

# Creating the bar plot from the DataFrame
ax = df.plot(kind='bar', figsize=(10, 8), width=0.8, color=['green', 'blue', 'red', 'black', 'purple'])

# Adding grid with major lines in purple and minor lines in grey
ax.grid(True, linestyle='-', linewidth=0.5, color='purple', which='major', axis='y')
ax.grid(True, linestyle=':', linewidth=0.5, color='grey', which='minor', axis='y')
ax.minorticks_on()

# Customizing the ticks
plt.tick_params(which='both', width=0.5)
plt.tick_params(which='major', length=7)
plt.tick_params(which='minor', length=4, color='grey')

# Adding title and axis labels
plt.title('Popularity of Programming Languages')
plt.xlabel('Index')
plt.ylabel('Values')

# Show legend
plt.legend(title='Legend', bbox_to_anchor=(1.05, 1), loc='upper left')

# Setting the x-axis labels to match the index of the DataFrame
ax.set_xticklabels(df.index)

# Displaying the plot
plt.show()
```

```python
# Task 6

# Given data
data = {
    'Country': ['United States', 'Great Britain', 'China', 'Russia', 'Germany'],
    'Gold Medals': [46, 27, 26, 19, 17]
}

# Creating a DataFrame
df = pd.DataFrame(data)

# Creating the pie chart
colors = ['blue', 'green', 'orange', 'red', 'brown']  # Custom colors for each slice
explode = (0.1, 0, 0, 0, 0)  # 'Explode' the 1st slice (United States)

plt.figure(figsize=(8, 8))
plt.pie(df['Gold Medals'], labels=df['Country'], autopct='%1.1f%%', startangle=140, colors=colors, explode=explode, shadow = True)

plt.title('Gold medal achievements of five most successful countries in 2016 Summer Olympics')
plt.show()
```
